### Taylor
We can Taylor expand about the 
$$
    E(a) \approx E(a_0) + \frac{\partial^2 E}{\partial a^2}(a-a_0)^2
$$

```python
def taylor(x, x0, a, b):
    return b + a*(x-x0)**2
```

```python
popt_taylor, pcov_taylor = curve_fit(taylor, alat, etot)
```
