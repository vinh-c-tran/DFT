# Silicon Density of States
1. We can calculate the density of states.
2. This is done as follows
    - We start by running a scf calculation
    - Next, we run a non-scf calculation 
    - Lastly, we run a dos calculation
3. Then we can plot the subsequent data 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
si_dos = pd.read_csv('si.dos.csv', skiprows=1)
```

```python
si_dos
```

```python
si_dos_data = si_dos.to_numpy()
```

```python
E = si_dos_data.transpose()[1]
dos_E = si_dos_data.transpose()[2]
int_dos_E = si_dos_data.transpose()[3]
```

```python
plt.plot(E, dos_E)
plt.grid() 
plt.xlabel('$(E-E_F)$ in eV')
plt.ylabel('Density of States')
plt.title('Silicon DOS')
plt.show()
```
