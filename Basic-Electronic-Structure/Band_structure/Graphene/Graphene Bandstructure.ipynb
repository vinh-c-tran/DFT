```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import subprocess

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
gp_band = pd.read_csv('bands.out.gnu', delim_whitespace = True, header = None, names = ['k', 'Energy'])
```

```python
gp_band
```

```python
gp = gp_band.to_numpy()
k_space = gp.transpose()[0]
energy = gp.transpose()[1]
```

## Plotting

```python
plt.scatter(k_space, energy)
plt.xlabel('$k$ space')
plt.ylabel('Energy in eV')
plt.title('Graphene')
plt.grid()
```
